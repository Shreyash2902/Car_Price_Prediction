# **Employment Status Prediction Using Logistic Regression & Decision Tree Classifiers**

# **Data Preprocessing**

```python
df.drop(["IND CODE NIC"],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

## **Feature Encoding**

```python
df=pd.get_dummies(df,columns=['TECH EDU_LVL','SEX'])
```

```python
 df["Gen Education LVL"].value_counts()
```

```python
# Define the order manually
from sklearn.preprocessing import OrdinalEncoder
order = [["illiterate","literate without formal schooling","below primary","primary","middle","secondary","higher secondary","diploma","graduate","postgraduate and above"]]

encoder = OrdinalEncoder(categories=order)
df['Gen Education LVL'] = encoder.fit_transform(df[['Gen Education LVL']])
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le=LabelEncoder()
le.fit(df["STATUS CODE"])
df["STATUS CODE"]=le.transform(df["STATUS CODE"])
```

```python
df
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split

# Select the top features before splitting the data
X = df.drop("STATUS CODE",axis=1)
y = df["STATUS CODE"]

X_train,X_test,y_train,y_test=train_test_split(X, y,test_size=0.2,random_state=9,stratify=y)
```

## **Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

# **Handling Imbalance**

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.model_selection import train_test_split

ros = RandomOverSampler(random_state=42)
X_trainr, y_trainr = ros.fit_resample(X_train, y_train)
X_testr,y_testr = ros.fit_resample(X_test,y_test)
```

# **Logistic Regression**

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lr=LogisticRegression(penalty="l2", solver='lbfgs' , max_iter=100)
lr.fit(X_trainr,y_trainr)
```

```python
y_t_pred=lr.predict(X_trainr)
print(classification_report(y_trainr,y_t_pred))
```

```python
y_pred=lr.predict(X_testr)
print(classification_report(y_testr,y_pred))
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42,max_depth=300,class_weight="balanced")
dt.fit(X_trainr,y_trainr)
```

```python
y_t_pred=dt.predict(X_trainr)
print(classification_report(y_trainr,y_t_pred))
```

```python
y_pred=dt.predict(X_testr)
print(classification_report(y_testr,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_trainr, y_trainr)

```

```python
y_pred=model.predict(X_testr)
print(classification_report(y_testr,y_pred))
```
